### **ID3**                             


Decision Tree Algorithms: ID3

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydotplus
from IPython.display import Image

# Loading the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
class_names = iris.target_names

# Creating a decision tree classifier using the ID3 algorithm
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(X, y)

# Generating the decision tree visualization
dot_data = export_graphviz(dt, out_file=None, feature_names=feature_names, class_names=class_names, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
