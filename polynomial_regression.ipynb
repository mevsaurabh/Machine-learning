```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sample dataset
np.random.seed(0)
X = np.random.rand(100, 1)
Y = 2 + 3 * X**3 + 4 * X**7 + 5 * X**9 + np.random.randn(100, 1)

# Create the design matrix
X = np.hstack((np.ones((X.shape[0], 1)), X**3, X**7, X**9))

# Calculate the coefficients using the normal equation
beta_hat = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)

# Make predictions
y_pred = X.dot(beta_hat)

# Plot the data and the fitted line
plt.scatter(X[:, 1], Y, color='b')
plt.plot(np.sort(X[:, 1]), y_pred[np.argsort(X[:, 1])], color='k')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('polynomial Regression')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sample dataset
np.random.seed(0)
X = np.random.rand(100, 1)
Y = 2 + 3 * X**3 + 4 * X**7 + 5 * X**9 + np.random.randn(100, 1)

# Create the design matrix
X_design = np.hstack((np.ones((X.shape[0], 1)), X**3, X**7, X**9))

# Calculate the coefficients using the normal equation
beta_hat = np.linalg.inv(X_design.T.dot(X_design)).dot(X_design.T).dot(Y)

# Print the calculated coefficients
print(f"b0: {beta_hat[0]}")
print(f"b1: {beta_hat[1]}")
print(f"b2: {beta_hat[2]}")
print(f"b3: {beta_hat[3]}")

# Make predictions
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
x_range_design = np.hstack((np.ones((x_range.shape[0], 1)), x_range**3, x_range**7, x_range**9))
y_pred = x_range_design.dot(beta_hat)

# Plot the data and the fitted line
plt.scatter(X, Y, color='b')
plt.plot(x_range, y_pred, color='r')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')

plt.show()
# Calculate the residuals
y_pred = X_design.dot(beta_hat)
residuals = Y - y_pred

# Create a fitted vs residuals plot
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='k', linestyle='--')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Fitted vs Residuals Plot')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import plot_leverage_resid2

# Create a sample dataset
np.random.seed(0)
X = np.random.rand(100, 1)
Y = 2 + 3 * X**3 + 4 * X**7 + 5 * X**9 + np.random.randn(100, 1)

# Create the design matrix
X_design = np.hstack((np.ones((X.shape[0], 1)), X**3, X**7, X**9))

# Fit the model using statsmodels
X_design = sm.add_constant(X_design)
results = sm.OLS(Y, X_design).fit()

# Print the calculated coefficients
print(f"b0: {results.params[0]}")
print(f"b1: {results.params[1]}")
print(f"b2: {results.params[2]}")
print(f"b3: {results.params[3]}")

# Make predictions
x_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
x_range_design = np.hstack((np.ones((x_range.shape[0], 1)), x_range**3, x_range**7, x_range**9))
y_pred = x_range_design.dot(results.params)

# Plot the data and the fitted line
plt.scatter(X, Y, color='b')
plt.plot(x_range, y_pred, color='r')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression')

plt.show()

# Calculate the residuals
residuals = results.resid

# Create a fitted vs residuals plot
plt.scatter(results.fittedvalues, residuals)
plt.axhline(y=0, color='k', linestyle='--')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Fitted vs Residuals Plot')
plt.show()

# Create a QQ plot
stats.probplot(residuals.flatten(), dist="norm", plot=plt)
plt.title('QQ Plot')
plt.show()

# Create a residuals vs leverage plot
fig = plt.figure(figsize=(8,6))
plot_leverage_resid2(results, ax=fig.gca())
plt.title('Residuals vs Leverage Plot')
plt.show()
```
